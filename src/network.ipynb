```python
import numpy as np

# -- Activation function and its derivative --

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)
```

```python
# -- XOR dataset --
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([[0],
              [1],
              [1],
              [0]])
```

```python
# -- Network initialisation --

np.random.seed(42)

# Architecture: 2 inputs -> 2 hidden neurons -> 1 output
# W1 shape (2, 2): each column is the weights INTO one hidden neuron
# W2 shape (2, 1): weights into the single output neuron
W1 = np.random.rand(2, 2) * 0.5 # (input size, hidden size)
b1 = np.zeros((1, 2)) # 1 bias per hidden neuron
W2 = np.random.rand(2, 1) * 0.5 # (hidden size, output size)
b2 = np.zeros((1, 1)) # 1 bias for output neuron
```

```python
# -- Forward Pass --

def forward(X):
    # Layer 1: z1 = X @ W1 + b1, a1 = sigmoid(z1)
    # @ is matrix multiplication
    z1 = X @ W1 + b1
    a1 = sigmoid(z1)

    # Layer 2: z2 = a1 @ W2 + b2, a2 = sigmoid(z2)
    z2 = a1 @ W2 + b2
    a2 = sigmoid(z2)

    # Return everything - we'll need z1, a1, z2 during backpropagation
    return a2, (z1, a1, z2) 

predictions, cache = forward(X)
print("Predictions (untrained):", predictions.T)
print("Targets: ", y.T)
```

```python
# -- Loss function --
def binary_cross_entropy(preds, targets):
    # Clip predictions to avoid log(0)
    preds = np.clip(preds, 1e-10, 1 - 1e-10)
    return -np.mean(targets * np.log(preds) + (1 - targets) * np.log(1 - preds))
```
